```python
# If you are in Jupyter, run this first in its own cell:
# %gui qt

import sys
from pathlib import Path

sys.path.insert(0, str(Path("src").resolve()))

from qiskit_metal.toolbox_metal.import_export import load_metal_design
from qiskit_metal import MetalGUI
from qiskit_metal.renderers.renderer_gds.gds_renderer import QGDSRenderer

# 1) Load design from one JSON file
design = load_metal_design("example_chip_design.metal.json")

# 2) Show in GUI
gui = MetalGUI(design)
gui.rebuild()
gui.autoscale()

# 3) Export to GDS
# Your example JSON has enable_renderers=false, so create GDS renderer if needed
if "gds" in design.renderers:
    gds = design.renderers["gds"]
else:
    gds = QGDSRenderer(design)

status = gds.export_to_gds("example_chip_design.gds")
print("GDS export status:", status)   # 1 means success
```
