# Chain DSL 2x2 GUI Test

This notebook builds the 2x2 chain YAML (Hamiltonian/Circuit/Netlist/Geometry) and opens the Metal GUI.

If the Qt window does not appear, run `%gui qt` in a cell before the GUI cell.

```python
from pathlib import Path
import sys

repo = Path.cwd()
src = repo / "src"
if src.exists() and str(src) not in sys.path:
    sys.path.insert(0, str(src))

yaml_path = repo / "2x2_4qubit.chain.metal.yaml"

from qiskit_metal.toolbox_metal.design_dsl import build_design
design = build_design(yaml_path)
design
```

```python
from qiskit_metal import MetalGUI

try:
    from PySide6.QtWidgets import QApplication
except ModuleNotFoundError:
    from PySide2.QtWidgets import QApplication

app = QApplication.instance() or QApplication([])
gui = MetalGUI(design)
gui.rebuild()
gui.autoscale()
gui.main_window.show()
```
